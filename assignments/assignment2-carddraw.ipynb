# Assignment 2 - Deck of Cards API

This is an API that simulates dealing a deck of cards  

Write a program that "deals" (prints out) 5 cards  
First you need to shuffle  
Get the deck_id, with the deck_id you can get the cards  
From there you need to print the value and the suit of each card.  

Check if the user has drawn a pair, triple, straight, or all of the same suit and congratulate the user.

### Importing libraries

```python
import requests
from collections import Counter
```

### Shuffling cards

```python
# URL to create and shuffle a new deck
url_suffle = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"

# Sending request to API
response = requests.get(url_suffle)

# print(response.text) #test, checking API return

# Converting response to JSON format
data = response.json()

#Extracting deck ID from API response
deck_id = data["deck_id"] 

# Printing deck ID
print(f"Deck ID: {deck_id}")
```

```python
url_draw = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count=5"
response = requests.get(url_draw)
#print(response.text) #test print statement
cards = response.json()
drawn_cards = cards["cards"] #Extracting cards from API response
```

Printing the drawn cards and checking pair, triple, suit and straight

```python
# Lists to store value and suit
card_value = []
card_suit = []

# Print statement for user:
print("\n Your cards:")

# Looping throght cards and storing values
for card in drawn_cards:
    value = card["value"]
    suit = card["suit"]
    # Appending each values to defined lists
    card_value.append(value)
    card_suit.append(suit)
    # Printing drawn cards to user
    print(f"{value} of {suit}")

# Assigning numbers to symbols
symbols = {"JACK": 11, "QUEEN": 12, "KING": 13, "ACE": 14}

#Variable to store drawn cards
numbers = []

# Looping throught card values and substitute symbols for numbers
for value in card_value:
    if value in symbols:
        numbers.append(symbols[value])
    else:
        numbers.append(int(value))

# Checking from pairs, triples, suits, straight
# Counting values and suits from deck using Counter
value_counts = Counter(card_value)
suit_counts = Counter(card_suit)

# Checking pairs
if 2 in value_counts.values():
    print("Congratulations, you have drawn a pair!")

# Checking triples
if 3 in value_counts.values():
    print("Congratulations, you have drawn a triple!")

# Looking for same suit
if len(suit_counts) == 1:
#if 1 in suit_counts.values() and len(suit_counts) == 1:
    print("Congratulations, all cards have the same suit!")

# Looking for straight

#Sorting card numbers
v = sorted(set(numbers))

# Check if the cards are unique
if len(v) == 5:
    # Checking for a straight or a low ace straight
    if v == [2, 3, 4, 5, 14] or (v[-1] - v[0] == 4):
        print("Congratulations, you have a straight!")
```
